```python
from fastquant import backtest, get_stock_data
```

```python
import investpy as ip
from datetime import datetime
import pandas as pd
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
ticker = 'vvar3'
today = datetime.today().date().strftime('%d/%m/%Y')
```

```python
df = ip.get_stock_historical_data(ticker, 'brazil', from_date = '1/1/2020', to_date = today)
df = df.drop(columns='Currency')
```

```python
#df = get_stock_data('MSFT','2018-01-01','2020-12-31')
```

```python
df.head()
```

```python
res = backtest('macd', df, fast_period=[12,13,14], slow_period=26, signal_period=9, dir_period=10, commission=0.490, verbose=False,)
```

```python
res
```

```python
strats_opt = { 
    "smac": {"fast_period": range(5,30,5), "slow_period": range(30, 100, 10)}, 
    'macd': {'fast_period': 12, 'slow_period': 26, 'signal_period':9, 'sma_period':30, 'dir_period':10}
}

res_opt = backtest("multi", df, strats=strats_opt, verbose=False, init_cash=10000, commission=4.90/100) #, sort_by='sharperatio')
```

```python
res_opt = res_opt.sort_values('pnl', ascending=False)
res_opt.head()
```
