```python
import investpy as ip
import pandas as pd
import numpy as np
from datetime import datetime
from scipy.stats import gmean
import ta
import mplfinance as mpf
```

```python
ticker = 'PETR4'
today = datetime.today().date().strftime('%d/%m/%Y')
```

```python
data = ip.get_stock_historical_data(ticker, 'brazil', from_date = '1/1/2020', to_date = today)
data = data.drop(columns='Currency')
data.head()
```

```python
mpf.plot(data, type='candle', volume=True, mav=(10, 20, 50))
```

```python
data['HA_Open'] = (data.shift(1).Open + data.shift(1).Close)/2
data['HA_High'] = data.High
data['HA_Low'] = data.Low
data['HA_Close'] = (data.Open + data.High + data.Low + data.Close)/4
data['log_Volime'] = np.log(data.Volume) - min(np.log(data.Volume))
data.head()
```

```python
data = data.drop(columns=['Open','High','Low','Close','Volume'])
data.columns = ['Open','High','Low','Close','Volume']
data = data.dropna()
```

```python
data.head()
```

```python
mpf.plot(data, type='candle', volume=True, mav=(10, 20, 50))
```

```python
data['lower_shaddow'] = data.apply(lambda x: max(0, (min(x.Open, x.Close) - x.Low)), axis = 1)
data['upper_shaddow'] = data.apply(lambda x: max(0, x.High - max(x.Open, x.Close)), axis = 1)
data['ratio_shaddow'] = data.apply(lambda x: min(x.lower_shaddow, x.upper_shaddow)/max(x.lower_shaddow, x.upper_shaddow), axis=1)
```

```python
data.head(10)
```

```python
data['sma_5'] = data.Close.rolling(20).mean()
```

```python
data['rise_5'] = (data.Close - data.sma_5)
```

```python
data.Close.plot()
data.sma_5.plot(c='red')
data.rise_5.plot()
```
